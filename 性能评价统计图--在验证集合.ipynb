```python
import numpy as np
import matplotlib.pyplot as plt

##############################################################################
# 1. 准备数据：此处用随机数据进行演示
#   假设有两个域：源域 (source) 与目标域 (target)，各有特征与标签。
#   为了计算信息增益，通常需要离散型标签(如二分类)。这里简单随机生成。
##############################################################################
np.random.seed(0)

# 源域数据
source_feature = np.random.rand(200)  # [0,1) 区间随机数
source_labels = (source_feature > 0.5).astype(int)  # 简单的二分类标签

# 目标域数据
target_feature = np.random.rand(200)
target_labels = (target_feature > 0.3).astype(int)  # 另一种二分类规则

##############################################################################
# 2. 定义计算信息增益的函数
#   计算方式：IG = Entropy(parent) - [ (Nl/N)*Entropy(left) + (Nr/N)*Entropy(right) ]
#   - feature: 特征值数组
#   - labels: 对应的离散型标签 (0/1)
#   - threshold: 分割阈值
##############################################################################
def compute_entropy(probabilities):
    """ 计算熵，避免对 0 取对数时的数值问题 """
    eps = 1e-10
    return -np.sum(probabilities * np.log2(probabilities + eps))

def calculate_information_gain(feature, labels, threshold):
    """
    在给定阈值下，将数据分为左右两部分后，计算信息增益 (IG)。
    这里只针对二分类标签进行演示；若为回归任务可考虑使用方差减少等指标。
    """
    # 左、右子节点掩码
    left_mask = (feature <= threshold)
    right_mask = (feature > threshold)

    # 若其中一侧无数据，则不产生信息增益
    if left_mask.sum() == 0 or right_mask.sum() == 0:
        return 0.0

    # 父节点（当前样本集）的熵
    def entropy_of_subset(subset_labels):
        values, counts = np.unique(subset_labels, return_counts=True)
        probs = counts / counts.sum()
        return compute_entropy(probs)

    parent_entropy = entropy_of_subset(labels)

    left_entropy = entropy_of_subset(labels[left_mask])
    right_entropy = entropy_of_subset(labels[right_mask])

    # 计算加权子熵
    n_total = len(labels)
    n_left = left_mask.sum()
    n_right = right_mask.sum()
    weighted_entropy = (n_left / n_total) * left_entropy + (n_right / n_total) * right_entropy

    # 信息增益
    info_gain = parent_entropy - weighted_entropy
    return info_gain

##############################################################################
# 3. 计算不同阈值下的 IG，并绘图
##############################################################################
thresholds = np.linspace(0, 1, 50)  # 在 [0,1] 区间取 50 个阈值
ig_source_list = []
ig_target_list = []

for th in thresholds:
    # 分别计算源域与目标域的 IG
    ig_s = calculate_information_gain(source_feature, source_labels, th)
    ig_t = calculate_information_gain(target_feature, target_labels, th)
    ig_source_list.append(ig_s)
    ig_target_list.append(ig_t)

##############################################################################
# 4. 绘制图像
##############################################################################
plt.figure(figsize=(6, 4),dpi=600)
plt.plot(thresholds, ig_source_list, color='black', label='source',linewidth=2)
plt.plot(thresholds, ig_target_list, color='red', label='target',linewidth=2)
# 设置坐标轴标签和标题，加粗，并设置为 Times New Roman 字体
plt.xlabel('Threshold', fontsize=14, fontweight='bold', family='Times New Roman')
plt.ylabel('Information Gain', fontsize=14, fontweight='bold', family='Times New Roman')
plt.title('IG value for thresholds', fontsize=16, fontweight='bold', family='Times New Roman')

# 设置坐标轴刻度标签的字体大小、加粗和字体
plt.xticks(fontsize=12, fontweight='bold', family='Times New Roman')
plt.yticks(fontsize=12, fontweight='bold', family='Times New Roman')

# 设置图例，字体加粗，字体为 Times New Roman
plt.legend(fontsize=12, loc='upper right', family='Times New Roman', frameon=True, framealpha=0.5)

# 增加图形外边框加粗
ax = plt.gca()
ax.spines['top'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)

# 显示网格
plt.grid(True)

# 调整布局并显示图像
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import jensenshannon
from scipy.interpolate import make_interp_spline

# 计算熵
def compute_entropy(probabilities):
    """ 计算熵，避免对 0 取对数时的数值问题 """
    eps = 1e-10
    return -np.sum(probabilities * np.log2(probabilities + eps))

# 计算信息增益
def calculate_information_gain(feature_values, labels, threshold):
    left_mask = feature_values <= threshold
    right_mask = feature_values > threshold

    if np.sum(left_mask) == 0 or np.sum(right_mask) == 0:
        return 0.0

    def entropy_of_subset(subset_labels):
        values, counts = np.unique(subset_labels, return_counts=True)
        probs = counts / counts.sum()
        return compute_entropy(probs)

    parent_entropy = entropy_of_subset(labels)
    left_entropy = entropy_of_subset(labels[left_mask])
    right_entropy = entropy_of_subset(labels[right_mask])

    n_total = len(labels)
    n_left = left_mask.sum()
    n_right = right_mask.sum()
    weighted_entropy = (n_left / n_total) * left_entropy + (n_right / n_total) * right_entropy

    return parent_entropy - weighted_entropy

# 计算Jensen-Shannon散度
def calculate_jensen_shannon_divergence(source_values, target_values, threshold, bins=10):
    left_source = source_values[source_values <= threshold]
    right_source = source_values[source_values > threshold]
    left_target = target_values[target_values <= threshold]
    right_target = target_values[target_values > threshold]

    def compute_hist(data, bins):
        hist, _ = np.histogram(data, bins=bins, density=True)
        hist = hist + 1e-10  # 防止出现零值
        return hist / hist.sum()

    if len(left_source) == 0 or len(left_target) == 0:
        js_left = 1.0
    else:
        js_left = jensenshannon(compute_hist(left_source, bins), compute_hist(left_target, bins))
    if len(right_source) == 0 or len(right_target) == 0:
        js_right = 1.0
    else:
        js_right = jensenshannon(compute_hist(right_source, bins), compute_hist(right_target, bins))

    return (js_left + js_right) / 2

# 计算目标函数并优化阈值
def optimize_threshold_with_JS(source_feature, target_feature, source_labels, target_labels, thresholds):
    ig_list = []
    js_list = []
    objective_list = []

    for th in thresholds:
        ig_s = calculate_information_gain(source_feature, source_labels, th)
        ig_t = calculate_information_gain(target_feature, target_labels, th)

        # 计算JS散度
        js_s = calculate_jensen_shannon_divergence(source_feature, target_feature, th)

        # 目标函数 = 信息增益 - lambda * JS散度
        lambda_param = 1.0  # 可调整该参数
        objective = ig_s - lambda_param * js_s

        ig_list.append(ig_s)
        js_list.append(js_s)
        objective_list.append(objective)

    # 归一化数据，使其范围保持在 [0,1]
    ig_list = np.array(ig_list)
    js_list = np.array(js_list)
    objective_list = np.array(objective_list)

    ig_list = (ig_list - ig_list.min()) / (ig_list.max() - ig_list.min() + 1e-10)
    js_list = (js_list - js_list.min()) / (js_list.max() - js_list.min() + 1e-10)
    objective_list = (objective_list - objective_list.min()) / (objective_list.max() - objective_list.min() + 1e-10)

    # 寻找最大目标函数值对应的阈值
    max_objective_index = np.argmax(objective_list)
    optimal_threshold = thresholds[max_objective_index]
    return optimal_threshold, ig_list, js_list, objective_list

# 生成样本数据（随机）
np.random.seed(0)
source_feature = np.random.rand(200)
source_labels = (source_feature > 0.5).astype(int)

target_feature = np.random.rand(200)
target_labels = (target_feature > 0.3).astype(int)

# 定义阈值范围
thresholds = np.linspace(0, 1, 50)

# 优化阈值
optimal_threshold, ig_list, js_list, objective_list = optimize_threshold_with_JS(
    source_feature, target_feature, source_labels, target_labels, thresholds)

# 绘制结果图
plt.figure(figsize=(6, 4), dpi=600)

# 绘制信息增益、JS散度和目标函数
plt.plot(thresholds, ig_list, color='black', label='Information Gain', linewidth=2)
#plt.plot(thresholds, js_list, color='blue', label='JS Divergence', linewidth=2)
plt.plot(thresholds, objective_list, color='red', label='Objective Function', linewidth=2)

# 标出优化后的阈值
plt.axvline(x=optimal_threshold, color='red', linestyle='--', label=f'Optimal Threshold: {optimal_threshold:.2f}')

# 设置标题、坐标轴标签和图例
plt.title('Optimization of Threshold Using DG', fontsize=14, fontweight='bold', family='Times New Roman')
plt.xlabel('Threshold', fontsize=12, fontweight='bold', family='Times New Roman')
plt.ylabel('Value', fontsize=12, fontweight='bold', family='Times New Roman')
plt.legend(fontsize=12, loc='upper right', family='Times New Roman')

# 设置图形外边框加粗
ax = plt.gca()
ax.spines['top'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)

# 设置坐标轴范围
plt.xlim(0, 1)
plt.ylim(0, 1)

# 显示网格
plt.grid(True)

# 调整布局并显示图像
plt.tight_layout()
plt.show()
```
