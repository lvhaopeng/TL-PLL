```python
import numpy as np
import matplotlib.pyplot as plt

# 数据
f1_1 = [0.89, 0.85, 0.75, 0.73]  # R²
f1_2 = [121.30, 116.8, 9.01, 3.8]  # RMSE
x = ['UTS', 'TYS', 'EL', 'RA']
x_ticks = np.arange(len(x))
width = 0.30  # 两组柱状图的宽度

# 创建图表和主轴
fig, ax1 = plt.subplots(figsize=(6, 4), dpi=600)

# 绘制 R² 的柱状图
bars1 = ax1.bar(x_ticks - width/2, f1_1, width, label='$R^2$', color='#8a9ac2', edgecolor='#000000', linewidth=1.5)
ax1.set_xlabel('Tensile property prediction', fontname='Times New Roman', fontsize=15, fontweight='bold')
ax1.set_ylabel('$R^2$', color='#000000', fontname='Times New Roman', fontsize=18, fontweight='bold')
ax1.tick_params(axis='y', labelcolor='#000000', labelsize=12, direction='in')
ax1.set_ylim(0, 1.2)
ax1.grid(axis='y', color='#000000', linestyle='--', linewidth=0.3)

# 在 R² 柱状图上添加数值标签
for bar in bars1:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, height + 0.02, f'{height:.2f}',
             ha='center', va='bottom', fontname='Times New Roman', fontsize=14, fontweight='bold')

# 创建共享 x 轴的第二个 y 轴
ax2 = ax1.twinx()

# 绘制 RMSE 的柱状图
bars2 = ax2.bar(x_ticks + width/2, f1_2, width, label='RMSE', color='#4367c4', edgecolor='#000000', linewidth=1.5)
ax2.set_ylabel('RMSE', color='#000000', fontname='Times New Roman', fontsize=15, fontweight='bold')
ax2.tick_params(axis='y', labelcolor='#000000', labelsize=15, direction='in')
ax2.set_ylim(0, max(f1_2) * 1.2)
plt.xticks(x_ticks, x, fontproperties='Times New Roman', fontsize=15)
plt.yticks(fontproperties='Times New Roman', fontsize=15)
plt.yticks(fontproperties='Times New Roman', fontsize=15, fontweight='bold')  # 加粗显示纵坐标标签
plt.ylim(0, 150)
ax1.set_ylim(0, 1.2)
# 在 RMSE 柱状图上添加数值标签
for bar in bars2:
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2, height + max(f1_2)*0.02, f'{height:.1f}',
             ha='center', va='bottom', fontname='Times New Roman', fontsize=14, fontweight='bold')

# 设置 x 轴刻度
ax1.set_xticks(x_ticks)
ax1.set_xticklabels(x, fontname='Times New Roman', fontsize=18, fontweight='bold')

# 添加图例
handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
plt.legend(handles1 + handles2, labels1 + labels2, loc='upper center', bbox_to_anchor=(0.6, 1.0),
           ncol=2, frameon=False, prop={'family': 'Times New Roman', 'size': 12, 'weight': 'bold'})

# 美化图表边框
for spine in ax1.spines.values():
    spine.set_linewidth(1.5)
    spine.set_color('black')
for spine in ax2.spines.values():
    spine.set_linewidth(1.5)
    spine.set_color('black')
# 确保所有 y 轴刻度标签使用 Times New Roman，大小和加粗
for label in ax1.get_yticklabels():
    label.set_fontname('Times New Roman')
    label.set_fontsize(15)
    label.set_fontweight('bold')

# 设置所有轴刻度线方向向内
ax1.tick_params(axis='both', direction='in', which='both')
ax2.tick_params(axis='both', direction='in', which='both')
# 确保所有文本使用 Times New Roman
plt.rc('font', family='Times New Roman')

# 调整布局
plt.tight_layout()

# 显示图表
plt.show()
```
